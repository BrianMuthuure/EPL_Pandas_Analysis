```python
import pandas as pd 
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
sns.set_style('darkgrid')
```

```python
df = pd.read_csv("EPL_20_21.csv")
```

```python
df
```

```python
df.drop(columns = ['xG', 'xA'], inplace=True)
```

```python
df
```

## Calculate Total Players - Per Club

```python
players_per_club = df['Club'].value_counts()
```

```python
players_per_club
```

```python
plt.figure(figsize=(12,10))
sns.barplot(players_per_club, players_per_club.index)
plt.xlabel("No. of Players")
```

## Total Goals Scored the Whole Season

```python
total_goals = df['Goals'].sum()
```

```python
total_goals
```

```python
total_penalties = df.Penalty_Goals.sum()
```

```python
total_penalties
```

```python
goals_no_penalties = total_goals - total_penalties
```

```python
goals_no_penalties
```

```python
plt.figure(figsize=(12,6))
slices = [goals_no_penalties, total_penalties]
labels = ['No Penalties', 'Penalties']
plt.pie(slices, labels=labels,  autopct='%1.0f%%');
```

## Players with the most goals

```python
highest_goals = df[['Name', 'Club', 'Goals', 'Assists']].sort_values(by='Goals', ascending=False).head(30)
```

## Players with most penalties

```python
highest_penalties = df[['Name', 'Club','Penalty_Attempted', 'Penalty_Goals']].sort_values(by='Penalty_Goals', ascending=False).head(20)
```

```python
highest_penalties 
```

## Players who have played many games

## Players who played many minutes
